# Skewness & Kurtosis

```python
sns.displot(data=df['C'], kind='kde')
```

```python
df['C'].skew()
```

```python
df['C'].kurtosis()
```

A value of 1.58  positively skewed:

1. The right tail (higher values) is longer.
2. Most data is concentrated on the left, but there are some large outliers pulling the tail to the right.

Kurtosis measures the tailedness (how heavy/light the tails are compared to a normal distribution).

1. Heavier tails than a normal distribution.
2. More data in the tails  more extreme outliers than you'd expect from a bell curve.
